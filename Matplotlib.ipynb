```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
y=2*x
y
```

# Line Plot 

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='r',linestyle=':',linewidth=2)         
plt.title("my first")
plt.xlabel("this is X axis")
plt.ylabel("this is Y-axis")
plt.show()
```

```python
#adding two lines in same plot

y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)  
plt.plot(x,y2,color='r',linestyle='-.',linewidth=3)
plt.title("TWO lines in plot")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)    ##True
plt.show()
```

```python
                                           ##ADDING SUB PLOT         IMPORTANT
plt.subplot(1,2,1)    #1row and 2collumn, 1 is subplot index
plt.plot(x,y1,color='b')
plt.title("1st subplot")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True) 

plt.subplot(1,2,2)     #2 at end is index
plt.plot(x,y2,color='g')
plt.title("2nd subplot")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)    ##True
plt.show()
```

```python
plt.subplot(2,1,1)    #2row and 1collumn, 1 is subplot index
plt.plot(x,y1,color='b')
plt.grid(True)
plt.subplot(2,1,2)     #2 at end is index
plt.plot(x,y2,color='g')
plt.show()
```

#                            Bar Plot         

```python
student={"carrie":80,'alex':60,"ryan":50}   
names=list(student.keys())         #SAVING IT IN LIST
values=list(student.values())
```

```python
names
```

```python
#plt.bar(names,values)
#plt.show()
```

```python
plt.bar(names,values)
plt.title("distribution of marks")
plt.xlabel("x axis- student name")
plt.ylabel("y axis- marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,values,color='r')    #use plt.barh  for horizantal
plt.title("marks")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.show()
```

```python
#SCATTER PLOT        its just a intersecn of pts
```

```python
x=[10,20,30,40,50,60,70]
a=[4,9,3,7,2,5,2]
plt.scatter(x,a)
plt.show()
```

```python
b=[8,2,6,8,1,5,4]
plt.scatter(x,a,color='r',marker='.',s=150)                           #2 scatter plt
plt.scatter(x,b,color='b',marker='*')
plt.show()
```

```python
#add subplotS OF SCATTERPLOT
x=[10,20,30,40,50,60,70]
a=[4,9,3,7,2,5,2]
b=[8,2,6,8,1,5,4]
plt.subplot(1,2,1)                          #always use sublot method for this
plt.scatter(x,a,c='r',marker='.',s=15)                          
plt.subplot(1,2,2)
plt.scatter(x,b,c='b',marker='*')
plt.show()
```
